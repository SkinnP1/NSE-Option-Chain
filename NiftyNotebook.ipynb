```python
import pandas as pd
import shutil
import time
import requests
import json
import numpy as np
from IPython.display import clear_output
from IPython.display import display
```

```python
start = 16500
end = 17500
```

```python
strikes = np.arange(start,end+50,50)
```

```python
df = pd.DataFrame(strikes, columns=['StrikePrice'])
df['Call-OI-Change'] = 0
df['Put-OI-Change'] = 0
df["Call-OI"] = 0
df['Put-OI'] = 0
```

```python
df
```

```python
first_price = 0
```

```python
    while True :
        new_url = 'https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY'
        headers = {'User-Agent': 'Mozilla/5.0'}
        page = requests.get(new_url,headers=headers)
        k = json.loads(page.text)
        ce = []
        pe = []
        for i in k['records']['data'] :
            if i['expiryDate'] == '23-Dec-2021' and i['strikePrice'] in strikes:
                ce.append(i['CE'])
                pe.append(i['PE'])
        spot_price = pe[0]['underlyingValue']
        
        ce_dataframe = pd.DataFrame(ce,columns=['strikePrice','openInterest'])
        pe_dataframe = pd.DataFrame(pe,columns=['strikePrice','openInterest'])
        ce_dataframe['Put-OI'] = pe_dataframe[['openInterest']]
        ce_dataframe['Call-OI-Change'] = 0
        ce_dataframe['Put-OI-Change'] = 0
        df_new = ce_dataframe.rename(columns = {'openInterest': 'Call-OI'}, inplace = False)
        df_answer = df_new.copy()
        ce_1 = df.loc[:,'Call-OI'].to_numpy()
        ce_2 = df_answer.loc[:,'Call-OI'].to_numpy()
        pe_1 = df.loc[:,'Put-OI'].to_numpy()
        pe_2 = df_answer.loc[:,'Put-OI'].to_numpy()
        ce_change = ce_2 - ce_1
        pe_change = pe_2 - pe_1
        df_answer['Call-OI-Change'] = pd.DataFrame(ce_change)
        df_answer['Put-OI-Change'] = pd.DataFrame(pe_change)
        df = df_answer.copy()
        clear_output()
        print("Spot Price = ", spot_price)
        print("Change Price = ", spot_price - first_price)
        first_price = spot_price
        display(df_answer)
        time.sleep(300)
```
